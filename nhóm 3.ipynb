```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('data.csv')
df.drop(['Dấu thời gian','Địa chỉ email','Điểm số'],axis=1,inplace=True)
df.dropna(inplace=True)
df
```

```python
df.describe()
```

```python
for col in df.columns:
    df[col] = df[col].str.split(':').apply(lambda x:x[0])
```

```python
df.columns = ['kha nang tra no cua khach hang','thu nhap','so tien vay','thu nhap con lai','so nam cong tac','so nguon thu nhap','tai san dam bao','do tuoi','trinh do']
df
```

```python
for col in df.columns:
    df[col] = df[col].astype('int')
```

```python
df['kha nang tra no cua khach hang'].value_counts()
```

```python
#EDA
import matplotlib.pyplot as plt
import seaborn as sns

def plot_stacked_bar(df,col,fontsize=11,figsize=(15,8),rot=0):
    temp = df.groupby([col,'kha nang tra no cua khach hang'])['kha nang tra no cua khach hang'].count().unstack(level=1)
    temp.plot(
        kind='bar',
        stacked=True,
        rot=rot,
        figsize=figsize,
        title=col,
        colormap='Pastel1'
    )
    plt.legend(['Tra no khong dung han','Tra no dung han'])
    plt.ylabel('So luong khach hang')
    plt.xlabel('')
    plt.show()
```

```python
X = df.drop('kha nang tra no cua khach hang',axis=1)
y = df['kha nang tra no cua khach hang']
# from imblearn.over_sampling import SMOTE
# strategy = {1:90,0:10}
# over = SMOTE(sampling_strategy=strategy,random_state=44)
# X,y = over.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=44,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=44)
logreg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('Train:',accuracy_score(y_train,logreg.predict(X_train)))
print('Test:',accuracy_score(y_test,logreg.predict(X_test)))
```

```python
confusion_matrix(y_train,logreg.predict(X_train))
```

```python
confusion_matrix(y_test,logreg.predict(X_test))
```

```python
print(classification_report(y_train,logreg.predict(X_train)))
```

```python
print(classification_report(y_test,logreg.predict(X_test)))
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print(f"R^2: {round(r2_score(y_test,logreg.predict(X_test)),2)}")
print(f"MAE: {round(mean_absolute_error(y_test,logreg.predict(X_test)),2)}")
print(f"RMSE: {round(mean_squared_error(y_test,logreg.predict(X_test), squared=False),2)}")
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_classifier = DecisionTreeClassifier()
DT_classifier.fit(X_train, y_train)

y_pred = DT_classifier.predict(X_test)
print(confusion_matrix(y_test,DT_classifier.predict(X_test)))
print(classification_report(y_test,DT_classifier.predict(X_test)))
print('Decision Tree accuracy: ', accuracy_score(y_test,DT_classifier.predict(X_test)))
```
